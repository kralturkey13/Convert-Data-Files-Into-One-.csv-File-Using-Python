<h2> Convert Data Files Into One .csv File Using Python</h2>

This program converts your seperate data files into one csv file. My files are like this:

 ![title](images/dat_files.png)

First, lets import necessary libraries.

```python
import numpy as np
import os
import pandas as pd
```

<p>"output_location" is where you want to save you file.</p>
<p>"file_name" is what you want your file name.</p>
<p>"folder_name" is the folder that contains your data files.</p>

```python
output_location="outputs/"
file_name="(Healthy) Motion Artifact Contaminated ECG Database"
folder_name="inputs/"
```

Let's create a variable and take the data files as list.

```python
files=os.listdir(folder_name)
print(files)
```

Now, create an empy list with a name "datas". Then open each data file with numpy and add to "datas" list.

```python
datas=[]
for f in range(len(files)):
    data = np.fromfile(folder_name + str(files[f]))
    datas.append(data)
print(datas)
```

Change the "datas" list into pandas DataFrame.

```python
df=pd.DataFrame(datas)
```

Save DataFrame as .csv file using numpy savetxt function.

```python
np.savetxt(output_location+str(file_name)+ ".csv",  
           df, 
           delimiter =", ",  
           fmt ='% s')
```

Our file should be saved into output location. Let's check the file is really in the output location.

```python
output_file=pd.read_csv(output_location + file_name +".csv")
output_file.head(26)
```
